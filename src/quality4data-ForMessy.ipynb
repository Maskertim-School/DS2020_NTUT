## 2. Data Quality Issue Function

```python
"""
data quality issues:
1. Uniqueness violation (col issue)
2. Missing Value (cell issue)
3. Outlier (cell issue)
4. Duplicated row (row issue)
5. Distinct Format (col issue)
constraint types:
1. explicit constraint
2. implicit constraint
metric:
conf: what probality that the issue is believed (implicit constraint)
prev: what proportion of datas in column that owns problem (explicit constraint)
"""

import math
import re
import copy

# list of the issues
ISSUE_LIST = ['uniqueness violation', 'distinct format', 'missing value', 'outlier', 'duplicated row']

# Uniqueness violation function (evaluated by prevalence(prev)) classified by dataset issue
# return prev, and column index
def eval_uniqueness(dataset, col_idx):
    ex_list = []  # record the existed value
    non_unique = [] # record duplicated
    prev = [1.0 for _ in range(len(dataset.columns))] # record the prev among these columns
    # if some ID is the same in ID_list, the ID is repeated
    for idx in col_idx:
        count = 0 # record how many datas are duplicated
        non_idx = []
        for i in range(len(dataset.iloc[:,idx])):
            if dataset.iloc[i,idx] not in ex_list:
                ex_list.append(dataset.iloc[i,idx])
            else:
                non_idx.append((i,idx))
                count = count+1
        non_unique.extend(non_idx)
        # calculate the prevalence
        prev[idx] = 1-round(count/len(dataset),3)
        ex_list.clear()
    return prev, col_idx, non_unique
               

# Missing Value function (evalutated by confidence) classified by single cell
# return list of idx of missing value (default of confidence 1), idx of miss format is [row, col]
def eval_missing(dataset, miss_token="?", col_nan=[], col_else=[]):
    miss_idx = [] # record index of missing value
    conf = [0.0 for _ in range(len(dataset.columns))] # record what probability that should not expected the missing value, and list indexed by column
    # calculate the confidence per cell
    for col_i in col_else:
        count = 0 # count the missing value in a column
        for row_j in range(len(dataset)):
            if dataset.iloc[row_j, col_i] == miss_token:
                miss_idx.append((row_j, col_i))
                count = count + 1
        conf[col_i] = round(count/len(dataset),3)
    # calculate for particular tag, nan
    for col_i in col_nan:
        count = 0 # count the missing value in a column
        for row_j in range(len(dataset)):
            if math.isnan(dataset.iloc[row_j, col_i]):
                miss_idx.append((row_j, col_i))
                count = count + 1
        conf[col_i] = round(count/len(dataset),3)
    return conf, miss_idx

# Outlier Function (evaluated by conf)
# return conf and outlier index
def eval_outlier(dataset, col_idx):
    out_idx = [] # record idx of outlier
    conf = [0.0 for _ in range(len(dataset.columns))] # record probability of confidence 
    # calculate the confidence per cell
    for col_i in col_idx:
        count = 0 # count the outlier in a column
        IQR = dataset.iloc[:,col_i].quantile(0.75)-dataset.iloc[:,col_i].quantile(0.25)
        lower_bound = dataset.iloc[:,col_i].quantile(0.25)-(IQR*1.5)
        upper_bound = dataset.iloc[:,col_i].quantile(0.75)+(IQR*1.5)
        for row_j in range(len(dataset)):
            if dataset.iloc[row_j, col_i] < lower_bound or dataset.iloc[row_j, col_i] > upper_bound:
                out_idx.append((row_j, col_i))
                count = count + 1
        conf[col_i] = round(count/len(dataset),3)
    return conf, out_idx

# Duplicated row (evalutated by confidence) classified by per row
# return conf and row index
def eval_duplicated(dataset, col_id):
    duplRow_idx = [] # record idx of duplicated row
    conf = 0.0 # record probability of confidence
    # calculate the confidence
    for i in range(len(dataset)):
        current = dataset.iloc[i,col_id]
        for j in range(i+1,len(dataset)):
            if current == dataset.iloc[j,col_id]:
                duplRow_idx.append(j)
    conf = round(len(duplRow_idx)/len(dataset),3)
    return conf, duplRow_idx

# Distinct format (evaluated by prevalence(prev)) classified by dataset issue
# return prev, and column index
def eval_distinctFormat(dataset, col_idx, isSame):
    distinct_idx = [] # record idx of distinct format
    prev = [1.0 for _ in range(len(dataset.columns))] # record how many datas are distinct
    # calculate the prevalence
    for col_i in col_idx:
        count = 0 # count the distinct in a column
        for i in range(len(dataset.iloc[:,col_i])):
            distinct_list = []
            if not isSame(dataset.iloc[:,col_i][i]):
                distinct_list.append(dataset.iloc[:,col_i][i])
                count = count+1
        distinct_idx.append(distinct_list)
        # calculate the prevalence
        prev[col_i] = 1 - round(count/len(dataset),3)
    return prev, col_idx, distinct_idx
```

### Test Data Quality Issue Functions

```python
'''
Test function whether is ok
'''

print('test eval_uniqueness-----')
prev_unique, unique_col, non_unique = eval_uniqueness(df, [0])
print('prev_unique:', prev_unique, 'unique cols:', unique_col, 'non_unique:',non_unique)

print('test eval_missing-----')
col_nan = [i for i in range(1, len(col_name))]
conf_miss, miss_list = eval_missing(df, "nan", col_nan, [0])
print('miss conf:', conf_miss, 'miss list [row, col]:', miss_list)
print('miss num:', len(miss_list))

print('test eval_outlier----')
col_out = [i for i in range(1, len(col_name))]
conf_out, out_idx = eval_outlier(df, col_out)
print('out_conf:', conf_out, 'out list:', out_idx)

print('test eval_duplicated----')
conf_dupli, dupli_rows = eval_duplicated(df, 0)
print('conf_dupli:', conf_dupli, 'duplicated rows:', dupli_rows)

print('test eval_distinct----')
prev_distinct, distinct_col, distinct_idx = eval_distinctFormat(df, [0],
                                            lambda data: re.compile('\d{4}[-/]\d{2}[-/]\d{2}').match(data))
print('prev_distinct:', prev_distinct, 'distinct cols:', distinct_col, 'distinct_idx:',distinct_idx)
```

## 3. Data Quality Score Function

```python
"""
Calculate the dataset quality score
"""
# store the info of issue (e.g., prev, conf, index list, etc.)
ROW_ISSUES = []
COLUMN_ISSUES = []
CELL_ISSUES = []

col = [i for i in range(1, len(col_name))]

# cell issue
conf_miss, miss_list = eval_missing(df, "nan", col, [0])
conf_out, out_idx = eval_outlier(df, col)
# row issue
conf_dupli, dupli_rows = eval_duplicated(df, 0)
# column issue
prev_distinct, distinct_col, distinct_idx = eval_distinctFormat(df, [0],
                                                  lambda data: re.compile('\d{4}[-/]\d{2}[-/]\d{2}').match(data))
prev_unique, unique_col, non_unique = eval_uniqueness(df, [0])

# add the info
ROW_ISSUES.append({'issue':'duplicated row','conf':conf_dupli, 'rows':dupli_rows})
COLUMN_ISSUES.append({'issue':'uniqueness violation','prev':prev_unique,'columns':unique_col,
                      'index':non_unique})
COLUMN_ISSUES.append({'issue':'distinct format','prev':prev_distinct,'columns':distinct_col,
                      'index':distinct_idx})
CELL_ISSUES.append({'issue':'missing value','conf':conf_miss,'cells':miss_list})
CELL_ISSUES.append({'issue':'outlier','conf':conf_out,'cells':out_idx})

# cal the dataset quality
def cal_quality(dataset):
    score = np.ones(dataset.shape)
    # cal row issue
    for ROW in ROW_ISSUES:
        for r in ROW['rows']:
            score[r,:] = score[r,:] * ROW['conf']
    # cal col issue
    for COL in COLUMN_ISSUES:
        for c in COL['columns']:
            score[:,c] = score[:,c] * COL['prev'][c]
    # cal cell issue
    for CELL in CELL_ISSUES:
        for cell in CELL['cells']:
            score[cell[0],cell[1]] = score[cell[0],cell[1]] * CELL['conf'][cell[1]]
    # cal final score
    SCORE_PER_COLUMN = copy.deepcopy(list(np.sum(score, axis=0)/len(dataset)))
    score = np.sum(np.sum(score, axis=0)/len(dataset))/dataset.shape[1]
    return score, SCORE_PER_COLUMN
score, SCORE_PER_COLUMN = cal_quality(df)
print('dataset quality score:',score)
```

## 4. Define the Quality Information in JSON format

```python
# record the info of issue
QUALITY_DATASET = []
# the info about: issue, type, index, count
# add lable
for issue in ISSUE_LIST:
    QUALITY_DATASET.append({'issue':issue, 'type':"none",'count':0})
# add type, index, count
count = 0
for i in range(len(COLUMN_ISSUES)):
    QUALITY_DATASET[count]['type'] = 'column'
    QUALITY_DATASET[count]['columns'] = list(COLUMN_ISSUES[i]['columns'])
    QUALITY_DATASET[count]['index'] = list(COLUMN_ISSUES[i]['index'])
    QUALITY_DATASET[count]['colCount'] = []
    for prev in COLUMN_ISSUES[i]['prev']:
        if prev != 1.0:
            QUALITY_DATASET[count]['colCount'].append(len(COLUMN_ISSUES[i]['index']))
        else:
            QUALITY_DATASET[count]['colCount'].append(0)
    QUALITY_DATASET[count]['count'] = int(np.sum(QUALITY_DATASET[count]['colCount']))
    QUALITY_DATASET[count]['prevalence'] = list(COLUMN_ISSUES[i]['prev'])
    count = count+1
for i in range(len(CELL_ISSUES)):
    QUALITY_DATASET[count]['type'] = 'cell'
    QUALITY_DATASET[count]['cells'] = list(CELL_ISSUES[i]['cells'])
    QUALITY_DATASET[count]['count'] = len(CELL_ISSUES[i]['cells'])
    QUALITY_DATASET[count]['conference'] = CELL_ISSUES[i]['conf']
    count = count+1
for i in range(len(ROW_ISSUES)):
    QUALITY_DATASET[count]['type'] = 'row'
    QUALITY_DATASET[count]['rows'] = list(ROW_ISSUES[i]['rows'])
    QUALITY_DATASET[count]['count'] = len(ROW_ISSUES[i]['rows'])
    QUALITY_DATASET[count]['conference'] = ROW_ISSUES[i]['conf']
# record the info of quality score
# the info about: id, colName, issues, sumOfProblem, score
SCORE_DATASET = []
# add id, colName
for i in range(len(df.columns)):
    SCORE_DATASET.append({'id':i,'colName':df_full_head.columns.values[i],
                         'issues':[],'score':0.0})
# add issues, sumOfProblem, score
for i in range(len(df.columns)):
    [SCORE_DATASET[i]['issues'].append(COLUMN_ISSUES[j]['issue']) for j in range(len(COLUMN_ISSUES)) 
                                  if COLUMN_ISSUES[j]['prev'][i] != 1]
    for j in range(len(CELL_ISSUES)):
        for cell in CELL_ISSUES[j]['cells']:
            if i == cell[1]:
                SCORE_DATASET[i]['issues'].append(CELL_ISSUES[j]['issue'])
                break
    for j in range(len(ROW_ISSUES)):
        if ROW_ISSUES[j]['rows']:
            SCORE_DATASET[i]['issues'].append(ROW_ISSUES[j]['issue'])
# add score
for i in range(len(df.columns)):
    SCORE_DATASET[i]['score'] = float(SCORE_PER_COLUMN[i])
```
